```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("data/cleaned_data.csv")

# Goals & Assists Distribution (Original)
plt.figure(figsize=(10, 5))
sns.histplot(df['Goals'], bins=20, kde=True, color='blue', label='Goals', alpha=0.6)
sns.histplot(df['Assists'], bins=20, kde=True, color='red', label='Assists', alpha=0.6)
plt.legend()
plt.title('Distribution of Goals and Assists')
plt.show()

# Histogram for Player Participation (Matches Played)
plt.figure(figsize=(8, 5))
sns.histplot(df["Matches"], bins=20, color="green", edgecolor="black")
plt.title("Distribution of Matches Played")
plt.xlabel("Matches")
plt.ylabel("Count")
plt.show()

# Pie Chart for Yellow Cards
plt.figure(figsize=(6, 6))
yellow_counts = df["Yellow_Cards"].value_counts()
plt.pie(yellow_counts, labels=yellow_counts.index, autopct='%1.1f%%', colors=["gold", "yellowgreen", "lightcoral"])
plt.title("Yellow Card Distribution")
plt.show()

# Pie Chart for Red Cards
plt.figure(figsize=(6, 6))
red_counts = df["Red_Cards"].value_counts()
plt.pie(red_counts, labels=red_counts.index, autopct='%1.1f%%', colors=["red", "salmon", "pink"])
plt.title("Red Card Distribution")
plt.show()
```
